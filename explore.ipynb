```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://en.wiktionary.org/wiki/Appendix:Roget's_thesaurus_classification")
html = r.text
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
classes = soup.find_all("h2")
for review_title in classes:
    ba_score_norm = review_title.findNext('span', {'class': 'mw-headline'})
    if ba_score_norm:  # Check if the element is found
        print(ba_score_norm.get_text())
```

```python
section = soup.find_all("h3")
for review_title in section:
    ba_score_norm = review_title.findNext('span', {'class': 'mw-headline'})
    if ba_score_norm:  # Check if the element is found
        print(ba_score_norm.get_text())
```

```python
# Find all <dl> elements
dl_elements = soup.find_all('dl')

# Process each <dl> element
for dl in dl_elements:
    # Extract all <dd> elements except the first
    dd_elements = dl.find_all('dd')[1:]

    # Process each <dd> element
    for dd in dd_elements:
        # Ignore <a> elements inside <small> tags
        small_tags = dd.find_all('small')
        for small in small_tags:
            small.decompose()

        # Find all <a> elements without a class attribute
        a_elements = dd.find_all('a', class_=False)

        # Extract text from each <a> element
        for a in a_elements:
            print(a.get_text())
```

```python
# Find all <dd> elements that contain <dl>
dd_elements = soup.find_all('dd')

# Process each <dd> element
for dd in dd_elements:
    if dd.find('dl'):
        # Extract and print the leading text (like "1. Being, in The Abstract")
        leading_text = dd.get_text().split('\n')[0]
        print(leading_text.strip())

        # Find all immediate <a> elements within nested <dl>
        nested_dl = dd.find('dl')
        nested_dd_elements = nested_dl.find_all('dd')

        for index, nested_dd in enumerate(nested_dd_elements, start=1):
            # Find the first <a> element
            a_element = nested_dd.find('a')
            if a_element:
                print(f"    {index}. {a_element.get_text()}")
```

```python
# Extract Classes
classes = soup.find_all("h2")
for class_title in classes:
    class_headline = class_title.findNext('span', {'class': 'mw-headline'})
    if class_headline:
        print(class_headline.get_text())

        # Iterate over sibling elements to find sections (h3) and divisions (dl)
        next_element = class_title.find_next_sibling()
        while next_element and next_element.name != "h2":
            if next_element.name == "h3":
                section_title = next_element.findNext('span', {'class': 'mw-headline'})
                if section_title:
                    print(f"  {section_title.get_text()}")

            elif next_element.name == 'dl':
                dd_elements = soup.find_all('dd')

                # Process each <dd> element
                for dd in dd_elements:
                    if dd.find('dl'):
                        # Extract and print the leading text (like "1. Being, in The Abstract")
                        leading_text = dd.get_text().split('\n')[0]
                        print(f"    {leading_text.strip()}")
                
                        # Find all immediate <a> elements within nested <dl>
                        nested_dl = dd.find('dl')
                        nested_dd_elements = nested_dl.find_all('dd')
                
                        for index, nested_dd in enumerate(nested_dd_elements, start=1):
                            # Find the first <a> element
                            a_element = nested_dd.find('a')
                            if a_element:
                                print(f"        {index}. {a_element.get_text()}")

            next_element = next_element.find_next_sibling()
```
